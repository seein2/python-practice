## 일주일

```python
import requests
import urllib
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import os
import re

weeks = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']

df = pd.DataFrame()

# 1주일치 7번 반복
for w in range(len(weeks)):
    base_url = f'https://m.comic.naver.com/webtoon/weekday?week={weeks[w]}'

    res = requests.get(base_url)
    html = res.text
    
    soup = BeautifulSoup(html, 'html.parser')

    # 각 요일별 데이터 추가
    webtoons = soup.select('ul.list_toon>.item>a.link') # 웹툰의 전체 정보
    titles = [] # 제목
    authors = [] # 작가
    count_nums = [] # 관심수
    days = [] # 요일
    for i in range(len(webtoons)):
        if webtoons[i].select_one('.title_text'): # 제목만 추출
            titles.append(webtoons[i].select_one('.title_text').text)
            days.append(weeks[w])
        if webtoons[i].select_one('.author'): # 작가만 추출
            authors.append(webtoons[i].select_one('.author').text.strip()) # strip()공백제거
        if webtoons[i].select_one('.count_num'): # 관심수만 추출
            count_nums.append(webtoons[i].select_one('.count_num').text)
        else:
            count_nums.append('NULL')


    # 썸네일 저장
    if not (os.path.isdir(f'thumbnail/{weeks[w]}')): # 썸네일/요일 폴더 검사(없으면 폴더 만들기)
        os.makedirs(os.path.join(f'thumbnail/{weeks[w]}'))
    
    img_path = []
    for k in range(len(soup.select('.link>.thumbnail>img'))):
        img_url = soup.select('.link>.thumbnail>img')[k]['src']
        img_name = soup.select('.link>.thumbnail>img')[k]['alt']
        img_name = re.sub('[^0-9a-zA-Zㄱ-힇]', '', img_name) # 특수기호 제거
        file_path = requests.get(img_url, headers={"User-Agent": "Mozilla/5.0"})

        img = open(f'thumbnail/{weeks[w]}/'+img_name+'.jpg', 'wb') # thumbnail파일/요일파일/'웹툰이름.jpg'
        img_path.append(f'thumbnail/{weeks[w]}/'+img_name+'.jpg') # 해당 이미지 위치주소 저장
        img.write(file_path.content) # img_url에서 받은 데이터를 img파일에 덮어쓰기
        img.close()
    dic = {
        '제목': titles,
        '작가': authors,
        '관심 수': count_nums,
        '요일': days,
        '경로': img_path,
    }
    temp_df = pd.DataFrame(dic)
    df = pd.concat([df, temp_df], ignore_index=True)
```

```python
from pytubefix import YouTube
from pytubefix.cli import on_progress

# driver = webdriver.Chrome()
# driver.get(youtube_url)

elm = driver.find_elements(By.XPATH, '//*[@id="thumbnail"]')
elm[2].click()

handles = driver.window_handles # 팝업을 선택할 수 있도록 만들어주는 속성
driver.switch_to.window(handles[0])

video = driver.find_element(By.CSS_SELECTOR, '#thumbnail > yt-image > img')
url = video.get_attribute('src')

yt = YouTube(url, on_complete_callback=on_progress)

ys = yt.streams.get_highest_resolution()
ys.download()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = sns.load_dataset("tips")

df_num = df.select_dtypes(include=['int', 'float'])
df_num['day'] = df['day']
df_num.groupby(['day']).mean()
print(df_num)

sns.barplot(x=df_num['day'], y=df_num['tip'], palette='RdBu')

plt.show()
```

```python
sns.boxplot(x=df_num['day'], y=df_num['total_bill'],hue=df['smoker'], palette='RdBu')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../실습자료/202407_202407_연령별인구현황_월간.csv', encoding='cp949', thousands=',')
df.iloc[[1, 9]]
```

```python
import re
p = re.compile('^2024년07월_남_\\d+|^2024년07월_여_\\d+')
cols = [c for c in df.columns if p.match(c)]
df[cols]

df_1 = df[['행정구역']+cols]
df_1
```

```python
age_people = [df_1.iloc[0,i] + df_1.iloc[0,i+11] for i in range(1,len(df_1.columns)//2+1)]
age_name = [f'{i}세' for i in range(0,110,10)]
```

```python
plt.rcParams["font.family"] = 'AppleGothic'
plt.barh(age_name, df_1.iloc[0,1:12].values*-1)
plt.barh(age_name, df_1.iloc[0,12:].values)

plt.show()
```

```python
df_1.iloc[1, 1:12].sum()
df_1.iloc[1, 12:].sum()

df_1.iloc[2, 1:12].sum()
df_1.iloc[2, 12:].sum()

plt.figure(figsize=(15,3))
plt.subplot(1,2,1)
plt.pie([df_1.iloc[1, 1:12].sum(), df_1.iloc[1, 12:].sum()], labels=['male', 'female'], autopct='%.1f%%', colors=['c', 'b'])
plt.legend()

plt.subplot(1,2,2)
plt.pie([df_1.iloc[2, 1:12].sum(), df_1.iloc[2, 12:].sum()], labels=['male', 'female'], autopct='%.1f%%', colors=['y', 'orange'])

plt.tight_layout()
plt.legend()
plt.show()
```

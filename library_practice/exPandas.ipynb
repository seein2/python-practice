```python
df = pd.read_csv('../실습자료/카페주문내역.csv', index_col=0)
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df
# 해당 날짜에 요일을 week컬럼 추가
df['week'] = df['date'].dt.dayofweek
df
# 토요일(5) 일요일(6) 이면 기존 price를 500원
df['price']=df.apply(lambda x : x['price']+500 if x['week'] > 4 else x['price'], axis=1)
df
```

```python
import seaborn as sns
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
anscombe['dataset'].value_counts()
```

```python
avg = anscombe.groupby('dataset').mean() # 평균
print(avg)
std = anscombe.groupby('dataset').std() # 표준편차
print(std)
var = anscombe.groupby('dataset').var() # 분산
print(var)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']

fig = plt.figure() # 그래프 기본 틀 생성
axes1 = fig.add_subplot(2, 2, 1) # row, col, index
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

axes1.plot(dataset_1['x'], dataset_1['y'], 'ro') # red circle
axes2.plot(dataset_2['x'], dataset_2['y'], 'go') # green circle
axes3.plot(dataset_3['x'], dataset_3['y'], 'bo') # blue circle
axes4.plot(dataset_4['x'], dataset_4['y'], 'yo') # yellow circle

axes1.set_title("dataset_1") # 각 데이터 제목 지정
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")
fig.suptitle("Anscombe Data") # 전체 데이터 제목 지정

fig.tight_layout() # 크기에 맞는 데이터 레이아웃 지정
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ids = pd.Series(['m01','m02','m03','m04'])
before = pd.Series([100, 110, 150, 90])
after = pd.Series([160, 130, 170, 120])
df = pd.concat([ids, before, after], axis=1)
df.columns = ['ID', 'Before', 'After']

plt.subplot(2,1,1)
plt.title('before')
plt.bar(df['ID'], df['Before'], width=0.8, color=['r', 'g', 'b', 'm'])

plt.subplot(2,1,2)
plt.title('after')
plt.barh(df['ID'], df['After'], height=0.8, color=['r', 'g', 'b', 'm'])

plt.tight_layout()
plt.show()
```

```python
plt.bar(df.index, df['Before'], width=0.5, color='yellow', align='center', label='before')
plt.bar(df.index+0.4, df['After'], width=0.5, color='orange', align='center', label='after')
plt.xticks(df.index, df['ID'])
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dic = {'city': ['Seoul','Incheon','Dejeon','Degu','Ulsan','Busan','Kawngju'],
       'lat':[37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16],
       'lon':[126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85],
       'p_dense': [16154, 2751, 2839, 2790, 1099, 4454, 2995]}
df = pd.DataFrame(dic)

plt.title('Dense')
plt.scatter(df['lon'], df['lat'], s=df['p_dense']*0.1, c=range(7), cmap='spring', alpha=0.5) # c 색, cmap , alpha 투명도
plt.xlabel('longitude')
plt.ylabel('latitude')

plt.text(df.iloc[0,2], df.iloc[0,1], df.iloc[0,0]) # plt.text(x좌표, y좌표, 글씨) : 그래프에 글씨 작성 
for i in range(len(df)):
    plt.text(df.iloc[i,2], df.iloc[i,1], df.iloc[i,0])
plt.show()
```
